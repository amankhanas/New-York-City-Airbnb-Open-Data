# Математико-статистический анализ
В этом ноутбуке рассчитываются описательные статистики, корреляции и взаимосвязи между признаками.

```python
import pandas as pd

df = pd.read_csv('../data/processed/airbnb_clean.csv')
numeric = df.select_dtypes(include=['int64', 'float64'])
desc = numeric.describe().T
corr = numeric.corr()

print('Descriptive statistics:')
display(desc)
```

```python
print('Correlation matrix:')
display(corr)
```

```python
price_corr = corr['price'].sort_values(ascending=False)
print('Correlation with price:')
display(price_corr)
```
